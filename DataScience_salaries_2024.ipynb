# Descriptive Analysis

```python
# What is the average, median, and mode of the salaries in the dataset?
average_salary = datascience_salaries_2024['salary'].mean()
median_salary = datascience_salaries_2024['salary'].median()
mode_salary = datascience_salaries_2024['salary'].mode()[0]
salary_range = (datascience_salaries_2024['salary'].min(),datascience_salaries_2024['salary'].max())
salary_std_dev = datascience_salaries_2024['salary'].std()

print("Average Salary:", average_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Salary Range:", salary_range)
print("Salary Standard Deviation:", salary_std_dev)

# Unique job title
unique_job_titles = datascience_salaries_2024['job_title'].nunique()
print("Number of Unique Job Titles:", unique_job_titles)

# Top 10 Most common job title
top_job_titles = datascience_salaries_2024['job_title'].value_counts().head(10)
print("Top 10 Job Titles:\n", top_job_titles)
```

```python
plt.figure(figsize=(10, 6))
bars = plt.bar(top_job_titles.index, top_job_titles.values, color='blue')

# Adding labels to the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

# Add title and labels
plt.title('Top 10 Most Common Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

# Exploratory Data Analysis(EDA)


```python
# Distribution of salaries
sns.histplot(datascience_salaries_2024['salary'], kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure that columns are properly named and stripped of any extra spaces
datascience_salaries_2024.columns = datascience_salaries_2024.columns.str.strip()

# Calculate the median salary for each job title and select the top 3
top_job_titles = datascience_salaries_2024.groupby('job_title')['salary'].median().nlargest(10).index

# Filter the DataFrame to include only these top job titles
filtered_data = datascience_salaries_2024[datascience_salaries_2024['job_title'].isin(top_job_titles)]

# Prepare data for the boxplot
salary_data = [filtered_data[filtered_data['job_title'] == title]['salary'] for title in top_job_titles]

# Plotting with Matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(salary_data, labels=top_job_titles)
plt.title('Salary by Job Title (Top 3)')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=90)

# Add labels for median salary on top of each boxplot
medians = [data.median() for data in salary_data]
for i, median in enumerate(medians):
    plt.text(i + 1, median + 1000, f'{median:.0f}', ha='center', color='black')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure that columns are properly named and stripped of any extra spaces
datascience_salaries_2024.columns = datascience_salaries_2024.columns.str.strip()

# Convert columns to appropriate data types if needed
datascience_salaries_2024['company_location'] = datascience_salaries_2024['company_location'].astype(str)
datascience_salaries_2024['salary'] = datascience_salaries_2024['salary'].astype(float)

# Calculate average salary by location
average_salary_by_location = datascience_salaries_2024.groupby('company_location')['salary'].mean()

# Get the top 5 locations by average salary
top_5_locations = average_salary_by_location.nlargest(5)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(top_5_locations.index, top_5_locations.values, color='skyblue')

# Add labels to each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:,.0f}', 
             ha='center', va='bottom', fontsize=10)

plt.title('Top 5 Locations by Average Salary')
plt.xlabel('Location')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)  # Rotate x-axis labels if needed for better readability
plt.show()
```
